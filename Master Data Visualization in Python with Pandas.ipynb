```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example DataFrame
data = {'Year': [2020, 2021, 2022, 2023], 'Sales': [200, 250, 300, 400]}
df = pd.DataFrame(data)
df
```

```python
# Line Plot
df.plot(x='Year', y='Sales', kind='line', title='Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Bar Plot
df.plot(x='Year', y='Sales', kind='bar', title='Yearly Sales (Bar Plot)')
plt.show()

# Area Plot
df.plot(x='Year', y='Sales', kind='area', title='Yearly Sales (Area Plot)', alpha=0.8)
plt.show()
```

```python
# Histogram
import numpy as np
df['RandomData'] = np.random.randn(4)
df['RandomData'].plot(kind='hist', bins=10, title='Data Distribution')
plt.show()
```

```python
# Scatter Plot
df.plot(x='Year', y='Sales', kind='scatter', title='Sales Trend')
plt.show()
```

```python
# Box Plot
df.boxplot(column='RandomData')
plt.show()
```

```python
df.plot(x='Year', y='Sales', kind='line')
plt.title('Yearly Sales Analysis')
plt.xlabel('Year')
plt.ylabel('Sales (in USD)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```
