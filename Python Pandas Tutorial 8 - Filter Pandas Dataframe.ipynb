# find rows of dataframe with year of 2007

```python
df[df['year']==2007]
```

# find rows of dataframe with year not in 2007

```python
df[df['year']!=2007]
```

# find rows of dataframe Whose Column Value is NOT NA/NAN

```python
df[df['year'].notnull()]
```

# find rows of dataframe based on a list?

```python
years = [2002, 2007]
```

```python
df[df['year'].isin(years)]
```

# find rows of dataframe based on values not in a list?

```python
df[~df['year'].isin(years)]
```
